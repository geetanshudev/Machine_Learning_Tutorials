# Decision Tree

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn import tree
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/luelhagos/Play-Tennis-Implementation-Using-Sklearn-Decision-Tree-Algorithm/master/Play%20Tennis.csv')
data
```

```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
data = data.apply(lbl.fit_transform)
data
```

```python
x = data.iloc[:,1:5]
y = data.iloc[:,5:6]
print(x.shape)
print(y.shape)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
ypred = dtc.predict(xtest)
print(ypred)
print(accuracy_score(ytest,ypred))
```

```python
cm = confusion_matrix(ytest,ypred)
```

```python
ConfusionMatrixDisplay(cm).plot()
```

```python
plot_tree(dtc)
```

```python
fig = plt.figure(figsize=(12,6))
plottree= tree.plot_tree(dtc,feature_names=x.columns,filled=True)
print(plottree)
```

# Underfitting in decision tree

1)max_depth is decrease or having small values then Underfitting occurs, 

2)min_sample_split is increases then underfitting happens

```python
dtc = DecisionTreeClassifier(criterion='entropy',splitter='random',max_depth=1,min_samples_split=9)
dtc.fit(xtrain,ytrain)
ypred = dtc.predict(xtest)
print(ypred)
print(accuracy_score(ytest,ypred))
print("\n")
fig = plt.figure(figsize=(6,6))
plottree= tree.plot_tree(dtc,feature_names=x.columns,filled=True)
print(plottree)
```

# Overfitting in decision tree

1) if max_depth have large values then overfitting happens or if it have value of None,

2) if min_sample_split have low values then overfitting occurs

```python
dtc = DecisionTreeClassifier(criterion='entropy',splitter='random',max_depth=10,min_samples_split=0.01)
dtc.fit(xtrain,ytrain)
ypred = dtc.predict(xtest)
print(ypred)
print(accuracy_score(ytest,ypred))
print("\n")
fig = plt.figure(figsize=(12,6))
plottree= tree.plot_tree(dtc,feature_names=x.columns,filled=True)
print(plottree)
print("Accuracy = ",accuracy_score(ytest,ypred))
```
