```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
import matplotlib.pyplot as plt
import time
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
```

```python
x = 6*np.random.rand(200,1)
y = 0.9*x**2+1.0*x+2+np.random.randn(200,1)
```

```python
plt.scatter(x,y,color='b')
plt.show()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
lr1 = LinearRegression()
lr1.fit(xtrain,ytrain)
ypred1 = lr1.predict(xtest)
r2_1 = r2_score(ytest,ypred1)
r2_1
```

```python
sgd=SGDRegressor()
sgd.fit(xtrain,ytrain)
ypred2 = sgd.predict(xtest)
r2_2 = r2_score(ytest,ypred2)
r2_2
```

```python
plt.scatter(x,y,color='b')
plt.plot(xtest,ypred1,color='red')
plt.show()
```

```python
plt.scatter(x,y,color='b')
plt.plot(xtest,ypred1,color='red',label='Linear Reg')
plt.plot(xtest,ypred2,color='orange',label='SGD Reg')
plt.legend()
plt.show()
```
