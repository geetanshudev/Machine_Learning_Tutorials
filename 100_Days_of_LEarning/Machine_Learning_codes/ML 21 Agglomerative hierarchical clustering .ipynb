```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('datasets/Placement.csv',usecols=['CGPA','IQ'])
data.sample(5)
```

```python
x = data.iloc[:,0]
y = data.iloc[:,1]
plt.figure(figsize=(10,7))
plt.scatter(x,y)
plt.show()
```

```python
dend = sch.dendrogram(sch.linkage(data,method='ward'))
dend
plt.figure(figsize=(10,12))
plt.show()
```

```python
aglclust = AgglomerativeClustering(n_clusters=3,affinity='l1',linkage='average')
label=aglclust.fit_predict(data)
label
```

```python
plt.scatter(x,y,c=aglclust.labels_,cmap='rainbow')
plt.show()
```
