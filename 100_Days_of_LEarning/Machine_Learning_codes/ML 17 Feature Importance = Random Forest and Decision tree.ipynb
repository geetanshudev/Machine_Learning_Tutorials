```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
data1 = pd.read_csv('MNIST/mnist_train.csv')
data1.sample(9)
```

```python
x = data1.iloc[:,1:]
y = data1.iloc[:,0]
sns.heatmap(x.iloc[12].values.reshape(28,28))
plt.show()
```

```python
for i in range(1,19):
    x = data1.iloc[:,1:]
    y = data1.iloc[:,0]
    sns.heatmap(x.iloc[i].values.reshape(28,28))
    plt.show()
```

```python
rf = RandomForestClassifier()
rf.fit(x,y)
rf.feature_importances_
```

```python
rf.estimators_[0].feature_importances_
for i,v in enumerate(rf.feature_importances_):
    print('Feature: %d,score:%.5f'%(i,v))
plt.bar([x for x in range(len(rf.feature_importances_))],rf.feature_importances_)
plt.show()
```
