```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
from sklearn.impute import SimpleImputer
```

# data 1

```python
data = pd.read_csv('datasets/MPG.csv',usecols=['mpg','horsepower','acceleration'])
impute = SimpleImputer(strategy='mean')
data = impute.fit_transform(data)
data = pd.DataFrame(data,columns=['mpg','horsepower','acceleration'])
data.isnull().sum()
```

```python
data.sample(6)
```

```python
x = data.iloc[:,0:1].values
y = data.iloc[:,2:3].values
plt.figure(figsize=(12,7))
plt.scatter(x,y)
plt.title("Simple graph")
plt.show()
```

```python
data = pd.read_csv('datasets/MPG.csv',usecols=['mpg','acceleration'])
data
```

```python
x = data.iloc[:,0:1]
y = data.iloc[:,1:2]
plt.figure(figsize=(9,6))
plt.scatter(x,y)
plt.title("MPg vs Acceleration")
plt.xlabel("MPG")
plt.ylabel("Accelaration")
plt.show()
```

```python
dbscan = DBSCAN(eps=2,min_samples=2)
cluster = dbscan.fit_predict(data)
cluster
```

```python
plt.figure(figsize=(9,6))
plt.xlabel("MPG")
plt.ylabel("Acceleration")
plt.title("DBSCAN Data Clustering")
plt.scatter(data.iloc[:,0],data.iloc[:,1],c=cluster,cmap='viridis')
plt.show()
```

# Data 2

```python
x,_ = make_circles(n_samples=500,factor=0.5,noise=.3,random_state=42)
dbscan1 = DBSCAN(eps=0.3,min_samples=3)
cluster1 = dbscan1.fit_predict(x)
```

```python
plt.figure(figsize=(9,4))
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.scatter(x[:,0],x[:,1],c=cluster1,cmap='viridis')

plt.show()
```
