```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Credit%20Default.csv")
data.sample(6)
```

# Method1



```python
x = data.iloc[:,0:4]
y = data.iloc[:,4:5]
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
knn = KNeighborsClassifier()
knn.fit(xtrain,ytrain)
ypred = knn.predict(xtest)
acc = accuracy_score(ypred,ytest)
print(acc*100)
```

# Method 2

```python
pca = PCA(n_components=2)
xtrain_pc = pca.fit_transform(xtrain)
xtest_pc = pca.transform(xtest)
```

```python
knn.fit(xtrain_pc,ytrain)
ypred1 = knn.predict(xtest_pc)
acc1 = accuracy_score(ypred1,ytest)
print(acc1*100)
```
