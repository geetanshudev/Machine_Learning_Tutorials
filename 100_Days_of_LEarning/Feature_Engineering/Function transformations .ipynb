```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import FunctionTransformer
```

```python
salary = pd.read_csv("C:\\Users\Geetanshu Dev\\Documents\\csv for Ml\\Adult Salary.csv",usecols=['Age','Gender','Education','NativeCountry'])
print(salary.shape)
salary.sample(5)
```

```python
salary.isnull().sum()
```

```python
x = salary['Age']
y = salary['Gender']

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.2,random_state=42)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
print(xtrain.skew())
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.hist(xtrain)
plt.title("Distribution Plot")

plt.subplot(122)
stats.probplot(xtrain,dist='norm',plot = plt)
plt.title('Q-Q plot')
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.hist(xtest)
plt.title("Distribution Plot")

plt.subplot(122)
stats.probplot(xtest,dist='norm',plot = plt)
plt.title('Q-Q plot')
plt.show()
```

```python
trf = FunctionTransformer(func=np.log1p)
xtrain_trans = trf.fit_transform(xtrain)
xtest_trans = trf.transform(xtest)
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.hist(xtrain_trans)
plt.title("Distribution Plot")

plt.subplot(122)
stats.probplot(xtrain_trans,dist='norm',plot = plt)
plt.title('Q-Q plot')
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.hist(xtest_trans)
plt.title("Distribution Plot")

plt.subplot(122)
stats.probplot(xtest_trans,dist='norm',plot = plt)
plt.title('Q-Q plot')
plt.show()
```
