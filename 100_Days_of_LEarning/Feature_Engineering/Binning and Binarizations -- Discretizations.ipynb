Converting numerical data to categorical data

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("titanic_dataset.csv")
data.sample(4)
x = data[['Age','Fare']]
```

```python
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest = train_test_split(x,test_size=0.2)
print(xtrain.shape)
print(xtest.shape)
```

```python
from sklearn.impute import SimpleImputer
smp = SimpleImputer()

xtrain_trans = smp.fit_transform(xtrain)
xtest_trans = smp.transform(xtest)
```

```python
print(xtrain_trans)
print("\n\n")
print(xtest_trans)
```

# Binning

we use sklearn's KBinsDiscretizer and try different strategy 

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
K_bin = KBinsDiscretizer(n_bins=10,encode='ordinal',strategy='quantile')
```

```python
xtrain_trs = K_bin.fit_transform(xtrain_trans)
xtest_trs = K_bin.transform(xtest_trans)
```

```python
print(xtrain_trs[0:5],"\n\n",xtrain_trans[0:5])
```
